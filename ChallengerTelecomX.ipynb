    Se os dados estiverem em formato de string JSON (ou seja, dentro de um dicionário ou variável), pode-se usar pd.json_normalize() diretamente.

    Inspeção dos Dados: Antes de qualquer transformação, é essencial entender a estrutura dos dados. Para isso, usamos:
    print(df.info()).
    Essas funções ajudam a identificar colunas aninhadas (dicionários dentro de colunas) e valores faltantes ou tipos de dados inconsistentes.

    Normalização dos Dados: Se algumas colunas contêm estruturas aninhadas (como dicionários dentro de colunas, comuns em JSON), usamos pd.json_normalize() para “achatar” esses dados em novas colunas.
    Como as
    df_customer = pd.json_normalize(df['customer'])
    df_account = pd.json_normalize(df['account'])
    df_phone = pd.json_normalize(df['phone'])
    df_internet = pd.json_normalize(df['internet'])

   Recomendações para Redução da Evasão

    Melhorar a experiência dos clientes com pagamento eletrônico:

        Muitos clientes que cancelaram utilizavam "Electronic check", sugerindo que esse método pode estar associado a dificuldades ou insatisfação.
        Aprimorar a interface de pagamento ou incentivar o uso de débito automático pode reduzir cancelamentos.

    Oferecer incentivos para contratos de longo prazo:

        A grande maioria dos que cancelaram estava no plano mensal, o que indica baixa fidelização.
        Ofereça descontos progressivos, bônus ou benefícios exclusivos para quem migra para planos anuais ou bienais.

    Monitorar e reduzir cobranças elevadas em segmentos específicos:

        O alto desvio padrão e o valor máximo indicam que alguns clientes podem estar pagando bem mais que a média.
        Analisar esses casos individualmente e aplicar políticas de revisão de cobrança pode evitar cancelamentos por custo.

    Campanhas de retenção direcionadas:

        Identificar clientes com alto risco de churn com base no perfil (método de pagamento, contrato mensal, cobrança alta) e oferecer:
